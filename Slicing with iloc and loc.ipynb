# Slicing with iloc and loc
### We'll be using a kaggle dataset of dog breeds (https://www.kaggle.com/kingburrito666/largest-dog-breed-data-set?select=2017.csv)

```python
import pandas as pd
import numpy as np
dogs = pd.read_csv('./2017.csv')
dogs.head()
```

### Is like slicing list but you need to sort the indexes on the dataframe


```python
dogs_srt = dogs.set_index(["Breed", "Color"]).sort_index()
print(dogs_srt)
```

### you use loc to slice through indexes, not values on the dataframe

```python
dogs_srt.loc["Chow Chow":"Poodle"]
```

### To slice in an inner lvl you need to do this

```python
dogs_srt.loc[("Labrador","Brown"):("Schnauzer","Grey")]
```

### To slice in inner lvl and columns you can do this:

```python
dogs_srt.loc[:,"DogName":"OwnerZip"]
```

```python
dogs_srt.loc[("Labrador","Brown"):("Schnauzer","Grey"),"DogName":"OwnerZip"]
```
