# Pivoting tables with pandas 

```python
import pandas as pd
import numpy as np
dogs = pd.read_csv('./2017.csv')
dogs.head()
```

### The pivots helps us to group data on a personalized way so that we can look at it in a perspective of two variables like in the example we use color and breed as our row and column indexes

```python
dogs_name_by_breed_vs_color = dogs.pivot_table("OwnerZip", index ="Breed", columns = "Color")
print(dogs_name_by_breed_vs_color)
```
